**Import libiraries:**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
```

**Load Dataset**

```python
file_path = "/content/drive/MyDrive/355227_6a050b3f72794_1778715455.xlsx"

df = pd.read_excel(file_path)
```

```python
print("First 5 Rows:")
display(df.head())

print("\nDataset Shape:")
print(df.shape)

print("\nDataset Info:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())
```

**Remove duplicate**

```python
duplicates_removed = df.duplicated().sum()

df = df.drop_duplicates()

print("\nDuplicates Removed:", duplicates_removed)
```

**Handle Missing Values**

```python
missing_before = df.isnull().sum().sum()

for col in df.columns:


    if df[col].dtype == 'object':

        mode_value = df[col].mode()[0]

        df[col] = df[col].fillna(mode_value)


    else:

        median_value = df[col].median()

        df[col] = df[col].fillna(median_value)

missing_after = df.isnull().sum().sum()

print("\nMissing Values Before:", missing_before)
print("Missing Values After:", missing_after)
```

**HANDLE OUTLIERS**

```python
age_col = "Age"

Q1 = df[age_col].quantile(0.25)
Q3 = df[age_col].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

before_outliers = len(df)

df = df[
    (df[age_col] >= lower_bound) &
    (df[age_col] <= upper_bound)
]

after_outliers = len(df)

outliers_removed = before_outliers - after_outliers

print("\nOutliers Removed:", outliers_removed)
```

**FEATURE ENGINEERING**

```python
text_columns = df.select_dtypes(include=['object']).columns

for col in text_columns:

    df[col] = (
        df[col]
        .astype(str)
        .str.strip()
        .str.lower()
    )

print("\nText Standardization Completed")
```

```python
# Encoding Categorical Data

label_encoders = {}

for col in text_columns:

    encoder = LabelEncoder()

    df[col] = encoder.fit_transform(df[col])

    label_encoders[col] = encoder

print("Encoding Completed")
```

```python
#SCALING
billing_col = "Monthly_Bill"

scaler = MinMaxScaler()

df[[billing_col]] = scaler.fit_transform(df[[billing_col]])

print("\nScaling Completed")
```

```python
print("\nFinal Dataset Shape:")
print(df.shape)

print("\nFirst 5 Rows After Cleaning:")
display(df.head())
```

```python
output_path = "/content/drive/MyDrive/cleaned_telecom_dataset.xlsx"

df.to_excel(output_path, index=False)

print("\nCleaned Dataset Saved Successfully")
```
